```python
def asignar_codigos_unicos(df, columna_codigos):
    codigos_numericos, codigos_unicos = pd.factorize(df[columna_codigos])
    df['Codigo_unico'] = codigos_numericos + 10_000_000
    return df
```

```python
df_0=asignar_codigos_unicos(df_0, 'Numero_Tarjeta')
df_1=asignar_codigos_unicos(df_1, 'Numero_Tarjeta')
df_2=asignar_codigos_unicos(df_2, 'Numero_Tarjeta')
df_3=asignar_codigos_unicos(df_3, 'Numero_Tarjeta')
df_4=asignar_codigos_unicos(df_4, 'Numero_Tarjeta')
df_total=asignar_codigos_unicos(df_total, 'Numero_Tarjeta')

max_valor0 = df_0['Codigo_unico'].max()
max_valor1 = df_1['Codigo_unico'].max()
max_valor2 = df_2['Codigo_unico'].max()
max_valor3 = df_3['Codigo_unico'].max()
max_valor4 = df_4['Codigo_unico'].max()
max_valortotal = df_total['Codigo_unico'].max()

print(f"El valor máximo en la columna 'Codigo_unico' es: {max_valor0}")
print(f"El valor máximo en la columna 'Codigo_unico' es: {max_valor1}")
print(f"El valor máximo en la columna 'Codigo_unico' es: {max_valor2}")
print(f"El valor máximo en la columna 'Codigo_unico' es: {max_valor3}")
print(f"El valor máximo en la columna 'Codigo_unico' es: {max_valor4}")
print(f"El valor máximo en la columna 'Codigo_unico' es: {max_valortotal}")
```
